```python
%pylab inline
#load the list of points
obs_data = loadtxt("movimiento.dat")
x_obs = obs_data[:,0]
y_obs = obs_data[:,1]
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return (-chi_squared)
```

```python
def my_model(x_obs, m,b,c):
    return m*x_obs**2 + (b*x_obs)+c
```

```python
m_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))
l_walk = empty((0))
c_walk= empty((0))

m_walk = append(m_walk, random.random())
b_walk = append(b_walk, random.random())
c_walk= append(c_walk,random.random())


y_init = my_model(x_obs, m_walk[0], b_walk[0],c_walk[0])
l_walk = append(l_walk, likelihood(y_obs, y_init))
print m_walk
print b_walk
print l_walk
print c_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    m_prime = np.random.normal(m_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)
    c_prime= np.random.normal(c_walk[i],0.1)

    y_init = my_model(x_obs, m_walk[i], b_walk[i],c_walk[i])
    y_prime = my_model(x_obs, m_prime, b_prime,c_prime)
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    alpha = (l_init/l_prime)
    if(alpha>=1.0):
        m_walk  = append(m_walk,m_prime)
        b_walk  = append(b_walk,b_prime)
        c_walk = append(c_walk,c_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta>=exp(-alpha)):
            m_walk = append(m_walk,m_prime)
            b_walk = append(b_walk,b_prime)
            c_walk = append (c_walk,c_prime)
            l_walk = append(l_walk, l_prime)
        else:
            m_walk = append(m_walk,m_walk[i])
            b_walk = append(b_walk,b_walk[i])
            c_walk = append(c_walk,c_walk[i])
            l_walk = append(l_walk, l_init)
```
