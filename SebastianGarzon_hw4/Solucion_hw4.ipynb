# 1. Filtrado datos 

Como se ha informado por parte de los estudiantes existe un numero aproximado de 100 datos que consisten en ruido y que deben ser separados de los datos que no contienen errores. Para ello se realizará un recorrido uno por uno en cada archivo para verificar que se trate de un comportamiento parabólico.
Por lo tanto en cada archivo se buscará el valor máximo y se evaluará el comportamiento de los puntos entre el 0 y el máximo dependiendo del tiempo. En un archivo sin errores todos los valores entre el 0 y el máximo serán mayores al anterior y menores al próximo. En los casos con archivos con ruido esta condicion no se cumplirá. Los archivos buenos serán copiados a una carpeta y los que contengan ruido serán enviados a otra.



```python
#metodo que cuenta los archivos en la carpeta Brahe-3141-f, incluye los 1000 datos y 1 archivo generado por python
dirlist = os.listdir("Brahe-3141-f")
numeroFiles= size(dirlist)
```

```python
#metodo que separa archivos buenos de los archivos con ruido, el recorrido inicia en el 1 ya que el primer nombre es un archivo generado por python
for i in range (numeroFiles-1000,numeroFiles):
    nombreArchivo=dirlist[i]
    f = open('Brahe-3141-f/'+nombreArchivo, 'r')
    matriz=np.loadtxt('Brahe-3141-f/'+nombreArchivo)
    posicionY=[]
    for j in range(0,38):
        posicionY.append(matriz[j][1])
    maximo=0
    contador=0
    #Metodo que saca el valor maximo de valores de Y
    for k in range (1,38):
        valor=posicionY[k]
        if valor>maximo:
            maximo=valor
            contador=contador+1
    #Verificador ,  metodo que cuenta el número de crcimientos desde el 0 hasta la posicion del valor maximo
    verificador=0
    for l in range(1,contador-1):
        if posicionY[l]<posicionY[l+1]:
            verificador+=1
        else:
            verificador+=0
    if not os.path.exists('archivosBuenos'):
        os.makedirs('archivosBuenos')
    if not os.path.exists('archivosRuido'):
        os.makedirs('archivosRuido')
    #Verificador, si el verificador es 2 numeros menor a la posicion en la que se encuentra el máximo quiere decir que todos los valores anteriores estan creciendo.
    if verificador==(contador-2):
       shutil.copy('Brahe-3141-f/'+nombreArchivo,'archivosBuenos/')
    else:
        shutil.copy('Brahe-3141-f/'+nombreArchivo,'archivosRuido/')
```

Como resultado tenemos la creación de 2 carpetas, en la carpeta 'archivosBuenos' se tienen 884 archivos mientras que en la carpeta 'archivosRuido' se tienen 116.
Los calculos para determinar gravedad, velocidad inicial y posicion inicial se determinarán únicamente con los 884 archivos de la carpeta 'archivosBuenos'

```python
dirlistBuenos=[]
dirlistBuenos = os.listdir("archivosBuenos/")
numeroBuenos=size(dirlistBuenos)
if os.path.exists('final_dat.dat')==True:
    os.remove('final_dat.dat')
for i in range (numeroBuenos-884,numeroBuenos):
    FD = open('final_dat.dat', 'a')
    nombreArchivo=dirlistBuenos[i]
    matriz2=np.loadtxt('archivosBuenos/'+nombreArchivo)
    posicionY=[]
    tiempo=[]
    resultados=[]
    string=''
    for j in range (0,38):
        posicionY.append(matriz2[j][1])
        tiempo.append(matriz2[j][0])
    resultados= polyfit(tiempo,posicionY,2)
    gravedad=str(resultados[0]*-2)
    velocidad=str(resultados[1])
    posicion_Inicial=str(resultados[2])
    theta_phi=re.findall("\d+.\d+", nombreArchivo)
    FD.write(theta_phi[0]+' '+theta_phi[1]+' '+gravedad+' '+velocidad+' '+posicion_Inicial+' '+'\n')
    FD.close()
```

# 3. Fluctuaciones de la gravedad con respecto a la media dependiendo del ángulo θ
